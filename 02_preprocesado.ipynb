```python
import pandas as pd
```

```python
datos = pd.read_csv("/content/train.csv", sep=",")
# Display de dataset
train = pd.DataFrame(datos)
train.head()
```

```python
train.info()
```

## Manejo de datos faltantes




```python
missing_values = train.isnull().sum()
print("Valores faltantes por columna:")
print(missing_values[missing_values > 0])
```

```python
# Impute missing values
for column in train.columns:
    if train[column].isnull().any():
        if train[column].dtype == 'object':
            # Impute categorical columns with mode
            mode_value = train[column].mode()[0]
            train[column].fillna(mode_value, inplace=True)
        elif train[column].dtype in ['int64', 'float64']:
            # Impute numerical columns with median
            median_value = train[column].median()
            train[column].fillna(median_value, inplace=True)

# Verify no remaining missing values
print("\nMissing values after imputation:")
print(train.isnull().sum())
```

## Convertir variables categoricas one-hot




```python
categorical_cols = train.select_dtypes(include='object').columns
train = pd.get_dummies(train, columns=categorical_cols, dummy_na=False)
display(train.head())
```

## Normalización


```python
numerical_cols = train.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_cols.remove('ID')
numerical_cols.remove('PERIODO_ACADEMICO')

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train[numerical_cols] = scaler.fit_transform(train[numerical_cols])

display(train.head())
```
